```python
# Dependencies and Setup
import pandas as pd

# Study data files
path2019 = "201907-citibike-tripdata.csv"
path2020 = "202007-citibike-tripdata.csv"

# Read the mouse data and the study results
data2019 = pd.read_csv(path2019)
data2020 = pd.read_csv(path2020)
```

```python
#adding year to the dataframes
data2019["year"]=2019
data2020["year"]=2020
```

```python
#sampling data to make it easier to read into tableau
data2020_sample=data2020.sample(frac=0.1)
data2019_sample=data2019.sample(frac=0.1)
```

```python
#concatenate tables
frames = [data2020_sample, data2019_sample]
data2019_2020 = pd.concat(frames)
```
